```python
# Exercicio Kart

import random # Biblioteca para gerar valores randomicos

#======================================================================================================================================
def cartRace(carts, laps): # Funo para gerar corrida
  raceInfo = [] # Criar lista de informaes da corrida
  
  for cart in carts:
    lap = 0
    time = 0
    totalTime = 0
    cartStatus = [] # Lista com todos os tempos por volta do cart
    while lap < laps: # Verificar se a volta atual  menor que o numero de voltas total
      lap += 1 
      time = random.uniform(10, 50) # Gerar numero float aleatrio de 10 a 50
      time = f"{time:.2f}" # Formatar o float
      cartStatus.append([lap, float(time)]) # Adicionar a volta e o tempo ao cartStatus

    for status in cartStatus: 
      totalTime += status[1] # Somar o tempo de todas as voltas no total

    totalTime = f"{totalTime:.2f}"
    run = [cart, cartStatus, float(totalTime)] # Todas as informaes de determinado cart na corrida
    raceInfo.append(run) # Adicionar as informaes de cada cart nas informaes da corrida 

  raceInfo.sort(key=lambda x: x[2], reverse=True) # Ordenar corredores de forma decrescente

  bestTime = 1000 # Variavel utilizada para descobrir o melhor tempo
  runners = [] # Lista que armazena o tempo total e o nome do corredor, utilizada para descobrir o vencedor
  bestLap = [0, 50] # Utilizada para descobrir a volta com o melhor tempo
  lapsInfo = [] #  Armazena as informaes de todas as voltas 
  timeSum = {} # Dicionario com a soma do tempo de cada volta

  for info in raceInfo:
    lapInfos = info[1] # Define lapInfos como todas as voltas dos corredores
    for lapInfo in lapInfos:
      lapsInfo.append(lapInfo)
      if lapInfo[1] < bestLap[1]:
        bestLap = [lapInfo[0], lapInfo[1], info[0]] # Armazena o nome, tempo e volta do corredor com o melhor tempo na volta

    if info[2] < bestTime:
      bestTime = info[2]
      runners.insert(0, [info[0],info[2]]) # Insere o melhor corredor no comeo da lista runners

  for lapRun, timeRun in lapsInfo: # Armazena a soma das voltas em um dicionario
    if lapRun in timeSum:
      timeSum[lapRun] += timeRun
    else:
      timeSum[lapRun] = timeRun

  sortedTimeSum = sorted(timeSum.items(), key=lambda x: x[1]) # Ordena o dicionario com as somas das voltas 
  averageBestLap = sortedTimeSum[0][1] / len(carts) # Pega a mdia da melhor volta
  averageBestLap = f"{averageBestLap: .2f}" # Formata a mdia
  bestLapInfos = [sortedTimeSum[0][0], averageBestLap] # Armazena a mdia e a volta em questo
  
  print()

  resultado = 0
  for runner in runners: # Estabelece as posies de cada corredor
    resultado += 1
    print(str(resultado) + ' Lugar:', runner[0], '-', str(runner[1]) + 's')

  
  print()
  print('A melhor volta foi feita em apenas', str(bestLap[1]) + 's, isso ocorreu na volta', bestLap[0], 'com o corredor', bestLap[2])
  print('A volta com a mdia mais baixa entre todas foi a volta', bestLapInfos[0], 'com uma mdia de' + bestLapInfos[1] + 's' )
#======================================================================================================================================

cartsOne = ['Lucas', 'Vinicius', 'Cabelo', 'Andre', 'Rafael', 'Felipe', 'Jos', 'Mario', 'Golden', 'Fabio']
cartsTwo = ['Lucas', 'Vinicius', 'Cabelo', 'Andre', 'Rafael', 'Felipe', 'Jos', 'Mario', 'Golden', 'Fabio', 'Joel', 'Rubinho', 'Hamilton', 'Foster', 'Giban', 'Carlos']
laps = 10
lapsEx = 25

cartRace(cartsOne, laps)
cartRace(cartsTwo, lapsEx)
```
